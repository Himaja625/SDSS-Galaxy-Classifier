**DATA PREPARATION**

**#Target Encoding**

```python
df['subclass'] = df['subclass'].map({'STARFORMING': 0, 'STARBURST': 1})
```

```python
df['subclass'].value_counts()
```

**#Handling Fake Missing Values**

```python
(df == -9999).sum()
```

*Converting all -9999 into proper NaN*

```python
df.replace(-9999, np.nan, inplace=True)
df.isnull().sum()
```

**#Filling These Missing Values**

```python
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
df.isnull().sum()
```

```python
#OUTLIER CHECK (IQR Method)
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum()
outliers
```

**SELECTING BEST FEATURES USING SELECT K BEST**

```python
#Separate Features & Target
X = df.drop(['subclass', 'class'], axis=1)
y = df['subclass']
```

```python
#Apply SelectKBest
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(X, y)
```

```python
#Get the Selected Feature Names
selected_features = X.columns[selector.get_support()]
selected_features
```

**BALANCING VALUES COUNTS USING SMOTE**

```python
#Creating New X and y using only Selected Features
X_selected = df[selected_features]
y = df['subclass']

print("Before SMOTE:")
print(y.value_counts())
```

```python
#Applying SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_selected, y)
```

```python
#Verifying Balancing
print("After SMOTE:")
print(y_balanced.value_counts())
```

**TRAIN–TEST SPLIT**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced,
    test_size=0.2,
    random_state=42,
    stratify=y_balanced
)

print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

**SCALING THE FEATURE VARIABLES USING "StandardScaler" METHOD**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaled Train Shape:", X_train_scaled.shape)
print("Scaled Test Shape:", X_test_scaled.shape)
```

**TRAINING THE MODEL IN MULTIPLE ALGORITHMS**

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create & Train Model
log_model = LogisticRegression(max_iter=800, solver='liblinear')
log_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_lr = log_model.predict(X_test_scaled)

# Results Summary Table
results_df = pd.DataFrame({
    "Metric" : ["Accuracy"],
    "Score"  : [accuracy_score(y_test, y_pred_lr)]
})

print("Logistic Regression Performance Summary")
display(results_df.style.set_properties(**{
    'background-color': '#f2e6ff',
    'color': '#3b0066',
    'font-weight': 'bold'
}))

# Classification Report
print("\n Detailed Classification Report:\n")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
plt.figure(figsize=(6,5))
cm = confusion_matrix(y_test, y_pred_lr)

sns.heatmap(
    cm, annot=True, fmt="d",
    cmap="Purples", linewidths=1, linecolor="black",
    cbar=False, annot_kws={"size":14}
)
plt.title("Confusion Matrix - Logistic Regression", fontsize=14)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Building the Model (Custom Parameters for Depth and Stability)
rf_model = RandomForestClassifier(
    n_estimators=180,
    max_depth=18,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42,
    bootstrap=True
)

# Train
rf_model.fit(X_train_scaled, y_train)

# Predict
y_pred_rf = rf_model.predict(X_test_scaled)

# Accuracy Table
rf_results = pd.DataFrame({
    "Metric": ["Accuracy"],
    "Score": [accuracy_score(y_test, y_pred_rf)]
})

print("Random Forest Performance Summary")
display(rf_results.style.set_properties(**{
    'background-color': '#e8f5e9',
    'color': '#1b5e20',
    'font-weight': 'bold'
}))

# Full Classification Report
print("\n Detailed Classification Report (Random Forest):\n")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix
plt.figure(figsize=(6,5))
cm = confusion_matrix(y_test, y_pred_rf)

sns.heatmap(
    cm, annot=True, fmt="d",
    cmap="Greens", linewidths=1, linecolor="black",
    cbar=False, annot_kws={"size":14}
)
plt.title("Confusion Matrix - Random Forest", fontsize=14)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Feature Importance Plot
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,6))
sns.barplot(
    x=importances[indices],
    y=X_selected.columns[indices],
    palette="crest"
)
plt.title("Feature Importance - Random Forest", fontsize=16)
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def decision_tree_performance(X_train, X_test, y_train, y_test):
    print("\n Training Decision Tree Classifier...")

    dt_clf = DecisionTreeClassifier(
        criterion='gini',
        max_depth=12,
        min_samples_split=6,
        min_samples_leaf=3,
        random_state=42
    )

    dt_clf.fit(X_train, y_train)

    # Predictions
    y_pred = dt_clf.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, y_pred)

    # Summary Table
    summary_df = pd.DataFrame({
        "Metric": ["Accuracy"],
        "Score": [round(acc, 6)]
    })

    print("\nDecision Tree Performance Summary")
    display(summary_df)

    print("\nDetailed Classification Report (Decision Tree):")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix Heatmap
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Greens", linewidths=.7,
                cbar=True, square=True)
    plt.title("Confusion Matrix - Decision Tree")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    return dt_clf, acc

# Running the Decision Tree
dt_model, dt_accuracy = decision_tree_performance(X_train_scaled, X_test_scaled, y_train, y_test)
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("\033[95m\n Training K-Nearest Neighbors Classifier...\033[0m")

#Train The Model
knn = KNeighborsClassifier(
    n_neighbors=7,
    weights='distance',
    metric='minkowski',
    p=2
)

knn.fit(X_train_scaled, y_train)

#Prediction
knn_pred = knn.predict(X_test_scaled)

#Metrics
knn_acc = accuracy_score(y_test, knn_pred)

print("\n\033[94mKNN Performance Summary\033[0m")
perf_df = pd.DataFrame({
    "Metric": ["Accuracy"],
    "Score": [knn_acc]
})
display(perf_df.style.background_gradient(cmap="Blues"))

print("\n\033[93mDetailed Classification Report (KNN):\033[0m")
print(classification_report(y_test, knn_pred))

#Confusion Matrix
cm_knn = confusion_matrix(y_test, knn_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm_knn, annot=True, cmap="Purples", fmt='g')
plt.title("Confusion Matrix - KNN", fontsize=14)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

#Distance Visualization
distances, _ = knn.kneighbors(X_test_scaled[:1])
plt.figure(figsize=(7,3))
plt.plot(distances[0], marker='o')
plt.title("Distance to Nearest Neighbors (Insight Plot)")
plt.xlabel("Neighbor Index")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
# Training Support Vector Machine (SVM)
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print("\nTraining Support Vector Machine (SVM)...\n")

svm_clf = SVC(kernel='rbf', C=1.5, gamma='scale')
svm_clf.fit(X_train_scaled, y_train)

svm_pred = svm_clf.predict(X_test_scaled)

svm_acc = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_acc)

print("\n Detailed Classification Report (SVM):\n")
print(classification_report(y_test, svm_pred))

# Confusion Matrix
cm_svm = confusion_matrix(y_test, svm_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm_svm, annot=True, fmt="d", cmap="viridis")
plt.title("Confusion Matrix - SVM")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Decision Confidence Plot
plt.figure(figsize=(8,4))
confidence = svm_clf.decision_function(X_test_scaled)
sns.histplot(confidence, bins=40, kde=True, color="purple")
plt.title("SVM Decision Confidence Distribution")
plt.xlabel("Decision Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Comparing Model Accuracies
import pandas as pd

# Retrieving accuracy scores from previously defined variables/DataFrames
log_reg_acc = results_df.loc[0, 'Score']
rf_acc = rf_results.loc[0, 'Score']

comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'Decision Tree', 'KNN', 'SVM'],
    'Accuracy': [log_reg_acc, rf_acc, dt_accuracy, knn_acc, svm_acc]
})

comparison = comparison.sort_values(by='Accuracy', ascending=False)
print("\n Model Accuracy Comparison:\n")
print(comparison)
```

```python
# Selecting the Best Model Automatically
best_row = comparison.iloc[0]
best_model_name = best_row['Model']
best_accuracy = best_row['Accuracy']

print(f"\n Best Performing Model: {best_model_name} ({best_accuracy:.4f})")

# Store the actual trained model reference
model_dict = {
    'Logistic Regression': log_model,
    'Random Forest': rf_model,
    'Decision Tree': dt_model,
    'KNN': knn,
    'SVM': svm_clf
}

best_model = model_dict[best_model_name]
```
